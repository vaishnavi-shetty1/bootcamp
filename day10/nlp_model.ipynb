```python
import re
```

```python
#cleaning
text='NLP, session is... going on!!'
clean = re.sub(r'[^a-zA-Z]','',text)
print(clean)
```

```python
#tokenization
text = 'The climate is beautiful'
tokens = text.split()
print(tokens)
```

```python
#stopword removal
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt_tab')

text= 'The climate is beautiful'
tokens = word_tokenize(text)
filtered = [word for word in tokens if word.lower() not in stopwords.words('english')]
print(filtered)
```

```python
#stemming
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
words = ['played','reading',]
for word in words:
  print(stemmer.stem(word))
```

```python
#lemmatization
#pos tagging
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
print(lemmatizer.lemmatize('running',pos='v'))
```

```python
#feature extraction
from sklearn.feature_extraction.text import CountVectorizer

vector = CountVectorizer()
text = ['The climate is beautiful','It is raining']
x = vector.fit_transform(text)
print(x.toarray())
print(vector.get_feature_names_out())
```

```python
#word embeddings
from gensim.models import Word2Vec
sentences = ['climate','beautiful'],['raining','climate']
model = Word2Vec(sentences,vector_size=20,window=2,min_count=1)
print(model.wv['climate'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

text = ['i love this phone',"i hate the phone's features"]
labels=[1,0]
vectorizer =  TfidfVectorizer()
x =vectorizer.fit_transform(text)
model = LogisticRegression()
model.fit(x,labels)

new_text = ['i like this phone']
new_x = vectorizer.transform(new_text)
predictions = model.predict(new_x)
print(predictions)
```
